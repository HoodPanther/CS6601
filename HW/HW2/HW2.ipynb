```python
def genMeans(ks, seed = None):
    random.seed(seed)
    means = random.sample(xrange(255), ks)
    return means

def getDists(image, means, dists):
    for i, mean in enumerate(means):
        dists[i] = image - mean
    return dists   

def getDist(mins):
    dist = np.sum(np.sum(np.square(mins)))
    return dist  
```
