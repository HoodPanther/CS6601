## 2) Implement: Gaussian Mixture Models and EM

```python
def logarithmicSumExponential(a, axis=None):
    #Compute the logarithmic sum of exponentials given a input
    #convert input into a np array
    a = np.asarray(a)
    #if a particular access is specified to sum over
    #then roll the axis so it is the first index
    if axis is None:
        a = a.ravel()
    else:
        a = rollaxis(a, axis)
    # Take the log of the sum of the exp of the input
    out = log(sum(exp(a), axis=0))
    return out
```

```python
class GMMandEM():
    def __init__(self, means_, vars_, weights_, n_components=1, thresh=1e-2, n_iter=100, n_init=1):
        # Store the initialized mean values
        self.means_ = means_
        # Store the initialized variance values
        self.vars_ = vars_
        # Store the initialized weights_ values
        self.weights_ = weights_
        # Store the initialized n_components values
        self.n_components = n_components
        # Define a threshold so that we can discern convergence
        self.thresh = thresh
        # Limit the number of iterations
        self.n_iter = n_iter
        # Limit the number of internal iterations
        self.n_init = n_init
        # Initialize the convergence flag to false
        self.converged_ = False
        
    def eval(self, X):
        # Compute the logarithmic profitability for every pixel in X
        lpr = log_normal_density(X, self.means_, self.vars_)
        # Tack on the logarithmic weights
        lpr += np.log(self.weights_)
        # Compute the logarithmic sum of exponentials of lpr
        logprob = logarithmicSumExponential(lpr, axis=1)
        # Acquire the responsibilities by using an exponential to convert from log space
        responsibilities = np.exp(lpr - logprob[:, np.newaxis])
        return logprob, responsibilities
    
    def expectation(self, X, log_likelihood):
        #Acquire the current likelihood and log format and responsibilities in exponential format
        curr_log_likelihood, responsibilities = self.eval(X)
        #Sum up the current likelihoods and append them to the growing list
        log_likelihood.append(curr_log_likelihood.sum())
        return log_likelihood, responsibilities
    
    def convergence_check(self, i, log_likelihood):
        # Check for convergence by first examining the loop index
        # make sure we are beyond our first iteration
        # then make sure the difference between the current likelihood and previous one
        # false below the specified threshold for convergence
        if i > 0 and abs(log_likelihood[-2] - log_likelihood[-1]) < self.thresh:
            # Modify the convergence flag
            self.converged_ = True
            # Notify that convergence has occurred
            return True
        else:
            # the Delta in iterative likelihood has 
            # not fallen below the convergence threshold
            return False
    
    def fit(self, X):
        ## Initialization step
        # Make sure X is an float np array
        X = np.asarray(X, dtype=np.float)
        # Reset the max recorded log prob to -infin
        max_log_prob = -np.infty

        ## Iterative Process
        # Iterate no more than the maximum specified iteration
        for _ in range(self.n_init):

            # Expectation Maximization algorithm
            log_likelihood = []
            # reset convergence flag to False
            self.converged_ = False
            # Iterate no more than the maximum specified iteration
            for i in xrange(self.n_iter):
                
                # Expectation step
                log_likelihood, responsibilities = self.expectation(X, log_likelihood)
                
                # Check for convergence.
                if self.convergence_check(i, log_likelihood):
                    break

                # Maximization step
                self.maximization(X, responsibilities, 'wmc')

            # Check to see if the results are better than the last
            # if so, keep them in some temp variables
            # this way if we reach out iteration limit,
            # will still have the best values we found
            if log_likelihood[-1] > max_log_prob:
                max_log_prob = log_likelihood[-1]
                best_params = {'weights': self.weights_,
                                'means': self.means_,
                                'covars': self.vars_}
                    
        #After convergence, store the best mean covariance and weights to remember later
        self.vars_ = best_params['covars']
        self.means_ = best_params['means']
        self.weights_ = best_params['weights']
        return self

    def maximization(self, X, responsibilities, params, min_covar=0):
        #Perform the maximization of the EM algorithm and return the weihgts
        weights = responsibilities.sum(axis=0)
        weighted_X_sum = np.dot(responsibilities.T, X)
        inverse_weights = 1.0 / (weights[:, np.newaxis])
        
        
        self.weights_ = weights / weights.sum()
        self.means_ = weighted_X_sum * inverse_weights
        self.vars_ = _var_mstep( self, X, responsibilities, weighted_X_sum, inverse_weights)
        return weights
    
    def compress(self, im):
        X = im.reshape((-1, 1))

        self.fit(X)
        curr_log_likelihood, responsibilities = self.eval(X)
        
        values = self.means_
        labels = responsibilities.argmax(axis=1)
        
        im_compressed = np.choose(labels, values)
        im_compressed.shape = im.shape
        score = logarithmicSumExponential(curr_log_likelihood)
        return im_compressed, score
    
    
def log_normal_density(X, means, covars):
    """Log probability for full covariance matrices."""
    from scipy import linalg
    import itertools
    
    solve_triangular = linalg.solve_triangular
        
    n_samples, n_dim = X.shape
    nmix = len(means)
    log_prob = np.empty((n_samples, nmix))
    for c, (mu, cv) in enumerate(itertools.izip(means, covars)):
        cv_chol = linalg.cholesky(cv, lower=True)
        
        cv_log_det = 2 * np.sum(np.log(np.diagonal(cv_chol)))
        cv_sol = solve_triangular(cv_chol, (X - mu).T, lower=True).T
        log_prob[:, c] = - .5 * (np.sum(cv_sol ** 2, axis=1) + n_dim * np.log(2 * np.pi) + cv_log_det)

    return log_prob
    
    
def _var_mstep(gmm, X, responsibilities, weighted_X_sum, norm):
    """Performing the variance M step"""
    # Eq. 12 from K. Murphy, "Fitting a Conditional Linear Gaussian Distribution"
    n_features = X.shape[1]
    cv = np.empty((gmm.n_components, n_features, n_features))
    for c in xrange(gmm.n_components):
        post = responsibilities[:, c]
        # Underflow Errors in doing post * X.T are  not important
        np.seterr(under='ignore')
        avg_cv = np.dot(post * X.T, X) / (post.sum())
        mu = gmm.means_[c][np.newaxis]
        cv[c] = avg_cv - np.dot(mu.T, mu)
    return cv
```

```python
def genMeans(n, seed = None):
    random.seed(seed)
    means = random.sample(xrange(255), n)
    return means
```

#### Using K-means as a better initialization algorithm

```python
def kMeans(im, n):
    X = im.reshape((-1, 1)) 
    k_means = cluster.KMeans(n)
    k_means.fit(X) 
    
    values = k_means.cluster_centers_.squeeze()
    labels = k_means.labels_
    
    im_compressed = np.choose(labels, values)
    im_compressed.shape = im.shape
    return im_compressed, values
```
